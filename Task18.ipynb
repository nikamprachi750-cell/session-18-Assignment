```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("ford_car_dataset.csv")              # 1. Load Dataset

print("Missing Values:")                              # 2. Handle Missing Values
print(df.isnull().sum())

df = df.drop_duplicates()                             # 3. Remove Duplicate Rows

                                                      # 4. EDA
plt.figure(figsize=(6,4))                             # Histogram - Price
plt.hist(df["price"], bins=20, edgecolor="black")
plt.title("Price Distribution")
plt.show()

plt.figure(figsize=(6,4))                           # Count Plot of Fuel Type
sns.countplot(x="fuelType", data=df)
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,6))                           # Correlation Heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

X = df.drop("price", axis=1)                          # 5. Input and Output Features
y = df["price"]

X = pd.get_dummies(X, columns=["model", "transmission", "fuelType"], dtype=int)       # 6. Encode Categorical Variables

numeric_cols = ["year", "mileage", "tax", "mpg", "engineSize"]                    # 7. Scale Numeric Features

scaler = StandardScaler()
X[numeric_cols] = scaler.fit_transform(X[numeric_cols])

print("First 5 Rows of Preprocessed Data:")                         # 8. Display Final Data
print(X.head())
```
